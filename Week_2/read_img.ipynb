```python
try:
    import urllib.request as urllib2
except ImportError:
    import urllib2
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/lualua0909/DeepLearning-LDS8/main/Data/depositphotos_6960117-stock-photo-stop-sign-isolated.jpg'
data = plt.imread(urllib2.urlopen(url), format='jpeg')
plt.imshow(data)
plt.show()
```
