```python
import time
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
cv2.__version__
```

```python
def convertToRGB(img):
  return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
test1 = cv2.imread('/content/drive/MyDrive/Cung cap Chapter9/FaceDetection_MTCNN/group_of_people.jpg')
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
```

```python
def detect_faces(cascade, img, scaleFactor):
  gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  plt.imshow(gray_img, cmap='gray')
  faces = cascade.detectMultiScale(gray_img, scaleFactor=scaleFactor, minNeighbors=5)
  print('Face found', len(faces))
  for (x, y, w, h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (255, 0, 255), 2)

  return img, len(faces)
```

```python
plt.imshow(gray_img, cmap='gray')
```

```python
haar_face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/haarcascade_frontalface_default.xml')
```

```python
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.05, minNeighbors=5)
```

* scaleFactor: tham số chỉ định kích thước hình ảnh  đc giảm bao nhiu ở mỗi tỉ lệ hình ảnh.
Ví dụ 1.03 là giảm kích thước xuống 3%.
* minNeighbors: chỉ định số lượng hàng xóm mà HCN ứng viên phải có. Giá trị càng cao thì dẫn đến ít phát hiện hơn nhưng chất lượng cao hơn. Default = 5

```python
print('Face found', len(faces))
```

```python
for (x, y, w, h) in faces:
  cv2.rectangle(test1, (x,y), (x+w, y+h), (255, 0, 255), 1)
```

```python
plt.imshow(convertToRGB(test1))
```
