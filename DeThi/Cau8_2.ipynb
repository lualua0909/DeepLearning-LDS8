```python
import time
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
cv2.__version__
```

```python
def detect_faces(cascade, img, scaleFactor):
  gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  plt.imshow(gray_img, cmap='gray')
  faces = cascade.detectMultiScale(gray_img, scaleFactor=scaleFactor, minNeighbors=3)
  print('Body found', len(faces))
  for (x, y, w, h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (255, 0, 255), 2)

  return img, len(faces)
```

```python
def convertToRGB(img):
  return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
test1 = cv2.imread('/content/drive/MyDrive/LDS8_DeThi_k270/Dataset/body.jpg')
```

```python
body_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/LDS8_K270_ONLINE_NguyenAnhDuy/haarcascade_fullbody.xml')
faces_detected_img, faces_found = detect_faces(body_cascade, test1, scaleFactor=1.3)
```

```python
plt.imshow(convertToRGB(faces_detected_img))
plt.axis('off')
```
