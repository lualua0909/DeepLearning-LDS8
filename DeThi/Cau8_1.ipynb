```python
import mtcnn
from mtcnn.mtcnn import MTCNN
import cv2
from skimage import io
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
image = io.imread('/content/drive/MyDrive/LDS8_DeThi_k270/Dataset/face_glasses_1.jpg')
plt.imshow(image)
```

```python
detector = MTCNN()
result = detector.detect_faces(image)
result
```

```python
bounding_box = result[0]['box']
keypoints = result[0]['keypoints']
```

```python
for p in result:
  bounding_box = p['box']
  keypoints = p['keypoints']

  cv2.rectangle(image, (bounding_box[0], bounding_box[1]),
                (bounding_box[0] + bounding_box[2], bounding_box[1] + bounding_box[3]),
                (255,0,0), 2)
  cv2.circle(image, (keypoints['left_eye']), 10, (0,255,0), 1)
  cv2.circle(image, (keypoints['right_eye']), 10, (0,255,0), 1)
```
